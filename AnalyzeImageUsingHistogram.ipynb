```python
# get image from link google image
import urllib.request
# from PIL import Image  # https://www.geeksforgeeks.org/python-pillow-a-fork-of-pil/
# https://media.geeksforgeeks.org/wp-content/uploads/20210318103632/gfg-300x300.png

urllib.request.urlretrieve(
    'https://img.freepik.com/free-photo/black-white-landscape-with-river_23-2149889411.jpg', 'lake.png'
), # link from google image

# img = Image.open("gfg.png") #load image filename
```

```python
# load image

import cv2
import matplotlib.pyplot as plt
import numpy as np

# Membaca gambar
image = cv2.imread('lake.png', cv2.IMREAD_GRAYSCALE)


# Dapatkan resolusi gambar dalam piksel
width, height = image.shape
print(f"Resolusi gambar: {width}x{height} piksel")

# Menampilkan gambar
plt.figure(figsize=(6,6))
plt.imshow(image, cmap='gray')
plt.title("Gambar Grayscale")
plt.show()
```

```python
# importing required libraries of opencv
import cv2

# importing library for plotting
from matplotlib import pyplot as plt

# reads an input image
img = cv2.imread('lake.png',0)

# find frequency of pixels in range 0-255
histr = cv2.calcHist([img],[0],None,[256],[0,256])

# show the plotting graph of an image
plt.plot(histr)
plt.show()
```

```python
import cv2
from matplotlib import pyplot as plt
img = cv2.imread('lake.png',0)

# alternative way to find histogram of an image
plt.hist(img.ravel(),256,[0,256])
plt.show()
```

EQUALIZATION

```python
equ = cv2.equalizeHist(img)
res = np.hstack((img,equ)) #stacking images side-by-side
cv2.imwrite('lake.png',res)
plt.imshow(res)
plt.show()
```

```python
# import OpenCV
import cv2

# import Numpy
import numpy as np

# import cv2_imshow from google.colab.patches
from google.colab.patches import cv2_imshow  # Import the correct function

# read an image using imread (grayscale mode)
image = cv2.imread('lake.png', cv2.IMREAD_GRAYSCALE)

# check if image is loaded properly
if image is None:
    print("Error: Gambar tidak ditemukan.")
else:
    # creating a Histogram Equalization of the image using cv2.equalizeHist()
    equ = cv2.equalizeHist(image)

    # stacking original and equalized images side-by-side
    res = np.hstack((image, equ))

    # show image input vs output using cv2_imshow instead of cv2.imshow
    cv2_imshow(res)  # Use cv2_imshow to display the image

    # wait for key event and close windows (These are not needed for cv2_imshow)
    # cv2.waitKey(0)
    # cv2.destroyAllWindows()
```
