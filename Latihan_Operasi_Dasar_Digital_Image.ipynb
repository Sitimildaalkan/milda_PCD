```python
# library
import urllib.request
import cv2 # https://docs.opencv.org/4.x/d6/d00/tutorial_py_root.html
import matplotlib.pyplot as plt
import numpy as np
```

```python
# get image from link google image

# first Image
urllib.request.urlretrieve(
'https://img.freepik.com/premium-photo/landscape-white-black_964616-3557.jpg', # link from google image
"gambar_1.png")

# Second Image
urllib.request.urlretrieve(
'https://i.pinimg.com/736x/33/fe/e9/33fee998d21e8d024c5de1260f722afb.jpg', # link from google image
"gambar_2.png")
```

```python
# load image
image_1 = cv2.imread('gambar_1.png',cv2.COLOR_BGR2GRAY) # cv2.IMREAD_GRAYSCALE
image_2 = cv2.imread('gambar_2.png',cv2.COLOR_BGR2GRAY)

# Dapatkan resolusi gambar dalam piksel
width, height,channel = image_1.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")

# Dapatkan resolusi gambar dalam piksel
width_2, height_2,channel_2 = image_2.shape
print(f"Resolusi gambar 2: {width_2}x{height_2}x{channel_2} piksel")
```

```python
# plot figure
plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.imshow(image_1)
plt.title("Gambar satu")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(image_2)
plt.title("Gambar dua")
plt.axis('off')

plt.show()
```

```python
image_1 = cv2.resize(image_1,(500,500))
image_2 = cv2.resize(image_2,(500,500))
```

```python
# hasil gambar Addition
result_add = cv2.add(image_1, image_2)
plt.imshow(result_add)
plt.show()

# Dapatkan resolusi gambar dalam piksel
width, height,channel = result_add.shape
print(f"Resolusi gambar 1: {width}x{height}x{channel} piksel")
```

```python
# Membandingkan data piksel
print("Image 1 Shape: ", image_1.shape)
print("Image 2 Shape: ", image_2.shape)
print("Result Image Shape: ", result_add.shape)

# Jika ingin melihat beberapa piksel untuk perbandingan, ambil subset data
image1_pixels = image_1[:5,:5,0]  # Ambil 5x5 bagian atas dari gambar 1 #channel blue
image2_pixels = image_2[:5,:5,0]  # Ambil 5x5 bagian atas dari gambar 2 #channel blue
image3_pixels = result_add[:5, :5, 0]  # Ambil 5x5 bagian atas dari gambar 2 #channel blue

print("\nData Piksel Gambar satu \n", image1_pixels)
print("\nData Piksel Gambar dua \n", image2_pixels)
print("\nData Piksel Gambar tiga \n", image3_pixels)
```

 Geometric Operation (rotation,translation,reflection)

```python
# prompt: please make  a code for rotation using open cv

import cv2
import matplotlib.pyplot as plt

# Load the image
img = cv2.imread('apple.png')

# Get the image dimensions
height, width = img.shape[:2]

# Define the rotation center
center = (width // 2, height // 2)

# Define the rotation angle (in degrees)
angle = 90

# Get the rotation matrix
rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)

# Perform the rotation
rotated_image = cv2.warpAffine(img, rotation_matrix, (width, height))

# Display the original and rotated images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))
plt.title('Rotation Image')
plt.axis('off')

plt.show()
```

```python
# prompt: please make a code for reflection using open cv

import urllib.request
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Download an image
urllib.request.urlretrieve(
    'https://i.pinimg.com/736x/33/fe/e9/33fee998d21e8d024c5de1260f722afb.jpg',
    'apple.png'
)

# Load the image
img = cv2.imread('apple.png')

# Define the reflection axis (x or y)
axis = 1  # 0 for x-axis, 1 for y-axis

# Perform reflection
if axis == 0:
  reflection_matrix = np.array([[-1, 0, img.shape[1]], [0, 1, 0]], dtype=np.float32)
  reflected_image = cv2.warpAffine(img, reflection_matrix, (img.shape[1], img.shape[0]))
elif axis == 1:
  reflection_matrix = np.array([[1, 0, 0], [0, -1, img.shape[0]]], dtype=np.float32)
  reflected_image = cv2.warpAffine(img, reflection_matrix, (img.shape[1], img.shape[0]))

# Display the original and reflected images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(reflected_image, cv2.COLOR_BGR2RGB))
plt.title('Reflection Image')
plt.axis('off')

plt.show()
```
